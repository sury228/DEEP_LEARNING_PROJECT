```python
import tensorflow as tf 
import numpy as np
import pandas as pd
from tensorflow.keras.models import load_model
import pickle
```

```python
model = load_model("churn_model.h5")

with open("scaler.pkl", "rb") as file:
    scaler = pickle.load(file)

with open("label_encoder.pkl", "rb") as file:
    label_encoder = pickle.load(file)

with open("onehot_encoder.pkl", "rb") as file:
    onehot_encoder = pickle.load(file)
```

```python
input_data = {
    'CreditScore': 600,
    'Geography': 'France',
    'Gender': 'Male',
    'Age': 40,
    'Tenure': 3,
    'Balance': 60000,
    'NumOfProducts': 2,
    'HasCrCard': 1,
    'IsActiveMember': 1,
    'EstimatedSalary': 50000
}
```

```python
input_df= pd.DataFrame([input_data])
```

```python
onehot_encoded_data = onehot_encoder.transform([[input_data['Geography']]]).toarray()
onehot_encoded_df = pd.DataFrame(onehot_encoded_data, columns=onehot_encoder.get_feature_names_out(['Geography']))
onehot_encoded_df
```

```python
input_df=pd.concat([input_df, onehot_encoded_df], axis=1)
input_df
```

```python
## encode the categorical features]
input_df['Gender'] = label_encoder.transform(input_df['Gender'])
input_df
```

```python
input_df.drop('Geography', axis=1, inplace=True)
input_df
```

```python
## scaling the data
scaled_input = scaler.transform(input_df)
scaled_input
```

```python
##predicting the output
prediction = model.predict(scaled_input)
prediction
```

```python
prediction_probability = prediction[0][0]
```

```python
if prediction_probability > 0.5:
    print(f"The customer is likely to churn with a probability of {prediction_probability:.2f}")
else:
    print(f"The customer is not likely to churn with a probability of {1 - prediction_probability:.2f}")
    
```
